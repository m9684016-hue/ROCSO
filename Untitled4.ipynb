```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem import WordNetLemmatizer, SnowballStemmer

# Descarga NLTK
nltk.download('punkt')
nltk.download('wordnet')
nltk.download("punkt_tab")

# Texto a tokenizar
text = 'Michael Jackson is singing a new son.'

print("tokenization: ")
print(sent_tokenize(text))

print('\nWords: ')
print(word_tokenize(text))

# NLP - Inicialización de lematizador y stemmer
lemmatizer = WordNetLemmatizer()
stemmer = SnowballStemmer("english")

# Palabras a analizar
words = ["singing", "sang", "sings"]

# Resultados - Formateo de tabla
print("\n{:<12} {:<12} {:<12}".format("Word", "Lemma", "Stem"))

for word in words:
    lemma = lemmatizer.lemmatize(word, pos='v')
    stem = stemmer.stem(word)
    print("{:<12} {:<12} {:<12}".format(word, lemma, stem))
```
